```python
# Filter numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Calculate correlation between numeric features and "churn"
correlation = numeric_columns.corrwith(df['churn']).sort_values(ascending=False)

# Print correlation results in order from strongest to weakest
print("Correlation with 'churn' (in descending order):")
print(correlation)
```

#### Check for instances of multicolinearity and drop the necessary columns

```python
# Filter numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Define a threshold for multicollinearity
threshold = 0.75

# Iterate through the correlation matrix and print pairs with multicollinearity
multicollinear_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            multicollinear_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j]))

print("Pairs with multicollinearity (correlation coefficient > {}):".format(threshold))
for pair in multicollinear_pairs:
    print(pair)
```

```python
# Define columns to drop based on multicollinear pairs and the state column since it is a not an integer.
columns_to_drop = ['state', 'number_vmail_messages', 'total_day_charge', 'total_eve_charge', 'total_night_charge', 'total_intl_charge']

# Drop the columns from the DataFrame
df_filtered = df.drop(columns=columns_to_drop)

# Print the new DataFrame to verify dropped columns
df_filtered.head(5)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Define the parameter grid
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create the decision tree classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5, scoring='accuracy')

# Perform hyperparameter tuning
grid_search.fit(X_train_scaled, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Instantiate the decision tree classifier with the best parameters
best_dt_classifier = DecisionTreeClassifier(**best_params, random_state=42)

# Train the model using the entire training set
best_dt_classifier.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_test = best_dt_classifier.predict(X_test_scaled)

# Evaluate the performance of the tuned model
accuracy_test = accuracy_score(y_test, y_pred_test)
print("Test Accuracy:", accuracy_test)

# Additional evaluation metrics
print("Classification Report:")
print(classification_report(y_test, y_pred_test))
# Plotting the confusion matrix
conf_mat_test = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat_test, annot=True, fmt="d", cmap="Blues", square=True, 
            xticklabels=['Predicted 0', 'Predicted 1'], 
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted labels')
plt.ylabel('Actual labels')
plt.title('Confusion Matrix (Test Set)')
plt.show()
```
