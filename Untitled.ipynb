### Data Cleaning ###

Based on the observations above on the dataset, the cleaning process will involve the following:
- Identify the target column for prediction, 'churn', which needs conversion from boolean values to 0s and 1s, where 1 denotes churn and 0 denotes no churn.
- Exclude the irrelevant part of the 'phone number' column by retaining only the area code.
- Replace spaces in column names with underscores for consistency and ease of reference during analysis.

```python
# Do away with the whitespaces in the column names and replace them with '_'
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Change the target variable 'churn' from True and False to 0s and 1s
df['churn'] = df['churn'].astype(int)
```

```python
# Print to see the changes
df.head(10)
```

Since both the State and Area Code can be used to determine a customer's location, the Area Code column can be dropped. 

```python
# Drop the state column
df = df.drop('area_code', axis=1)
```

The phone number column should be a unique identifier, since no two people should have the same phone number. Therefore, the following needs to be done: 
- change the data type to integer
- make it the index column
- check for duplicates and remove them

```python
# Change the data type to integer
df['phone_number'] = df['phone_number'].str.replace('-', '').astype(int)
```

```python
# Check for duplicates
duplicates = df.duplicated('phone_number')

# Show the duplicated rows
duplicated_rows = df[duplicates]
duplicated_rows
```

There are no duplicates.

```python
# Make it the index column 
df.set_index('phone_number', inplace=True)
```

## Exploratory Data Analysis ##

In the exploratory data analysis (EDA) section, attention is centered on understanding the dynamics of the "churn" variable within the dataset. By delving into the characteristics and patterns associated with churn, valuable insights are uncovered that can inform predictive modeling and retention strategies. Through visualizations, statistical summaries, and correlation analyses, factors influencing churn are identified and the potential relationships with other variables explored. 

```python
 # Calculate the counts of churn and no churn
churn_counts = df['churn'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(churn_counts, labels= ['No Churn', 'Churn'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Churn')
plt.axis('equal') 
plt.show()
```

The pie chart shows that only a small percentage (14.5%) of the customers are leaving, and the majority (85.5%) remain with the service provider. These percentages show the company's customer retention rate.

### Since the company's aim is to reduce the churn rate, the next step is to look into the data, to understand the distribution of customers. Therefore, the distribution of churn per state is explored below. ###

```python
# Group the data by state and churn status and counting the number of occurrences
churn_per_state = df.groupby(['state', 'churn']).size().unstack()

# Plot the bar chart
churn_per_state.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Customer distribution per State')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.legend(['No Churn', 'Churn'], title='Churn Status')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Focus only on churned customers
df_churn = df[df['churn'] == 1]  
# Find the customer count per state for those who churned
churn_count = df_churn['state'].value_counts()  

# Plot a bar graph for the number of churned customers per state
plt.figure(figsize=(15, 6))
churn_count.plot(kind='bar')
plt.title('Number of Churned Customers for each State')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

From the bar graph above, New Jersey had the most churned customers, while the state of Hawaii had the least. However, it is important to note that this shows the number of customers who churned, and not an indication of the rate of retention for the specified state. 

```python
# Find the top 5 states with the highest number of churned customers
top_5_churn_states = df_churn['state'].value_counts().head(5)

# Plot a bar graph for the number of churned customers per state
plt.figure(figsize=(12, 6))
top_5_churn_states.plot(kind='bar')
plt.title('Number of Churned Customers for Top 5 States')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=40)
plt.show()
```

Above graph shows the top 5 states with thew highest number of churned customers.

### Correlation of the different features ###

```python
# Calculate the correlation matrix for numerical features
corr_matrix = df.corr(numeric_only = True)

# Set the figure size for the heatmap
plt.figure(figsize=(16, 12))

# Create a heatmap with correlation values annotated
sns.heatmap(data=corr_matrix, annot=True, cmap='Blues')
```

The following are some of the observations that can be made from the table above:
- There's a weak (almost negligible) correlation between most of the features.
- There's a weak positive correlation between the target variable "churn" and the number of customer service calls, total day charges and total day minutes.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate churn rate percentage for each number of customer service calls
churn_rate = df.groupby('customer_service_calls')['churn'].mean() * 100

# Plotting a bar plot
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=churn_rate.index, y=churn_rate.values, hue=churn_rate.index, palette='Blues', legend=False)
ax.bar_label(ax.containers[0], fmt='%.1f%%', label_type='edge') # Add % labels

# Adding title and labels
plt.title('Rate of churning by Customer Service Calls as a percentage')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Churn Rate %')

# Display the plot
plt.show()
```

This shows that customers who call customer service alot, are more likely to churn than those who don't. The frequency of customer service calls demonstrates a proportional relationship with the likelihood of customers churning, with a noticeable surge in churn rates observed after the 6th call. This suggests a critical threshold where further interactions may negatively impact customer retention, potentially leading to increased churn rates.
